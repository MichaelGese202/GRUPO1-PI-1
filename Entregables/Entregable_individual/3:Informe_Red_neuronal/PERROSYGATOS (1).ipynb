## Visualización de imagenes

```python
import zipfile
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Ruta del zip descargado
zip_path = "/tmp/cats_and_dogs_filtered.zip"

# Extraer dataset
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/tmp")

# Rutas de imágenes
base_dir = "/tmp/cats_and_dogs_filtered"
train_dir = os.path.join(base_dir, "train")
train_cats_dir = os.path.join(train_dir, "cats")
train_dogs_dir = os.path.join(train_dir, "dogs")

# Seleccionar aleatoriamente 4 imágenes de cada clase
cat_images = random.sample(os.listdir(train_cats_dir), 4)
dog_images = random.sample(os.listdir(train_dogs_dir), 4)

# Crear figura con 2 filas (perros y gatos) y 4 columnas
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

# Primera fila: perros
for i, fname in enumerate(dog_images):
    img = mpimg.imread(os.path.join(train_dogs_dir, fname))
    axes[0, i].imshow(img)
    axes[0, i].set_title("Perro (Target = 1)")
    axes[0, i].axis("off")

# Segunda fila: gatos
for i, fname in enumerate(cat_images):
    img = mpimg.imread(os.path.join(train_cats_dir, fname))
    axes[1, i].imshow(img)
    axes[1, i].set_title("Gato (Target = 0)")
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()
```
